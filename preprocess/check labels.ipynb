```python
import os
import glob
import SimpleITK as sitk
import numpy as np
from tqdm import tqdm

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
all_files = sorted(glob.glob('/Users/rk588/data/lnq/intermediate/data/images/preprocessed/*.nrrd'))
label_files = sorted(glob.glob('/Users/rk588/data/lnq/intermediate/data/images/preprocessed/*_label.nrrd'))
```

```python
lymph_node_sizes = dict()
for label_file in tqdm(label_files):
    label = sitk.ReadImage(label_file)
    nda = sitk.GetArrayFromImage(label)
    lymph_node_sizes[os.path.basename(label_file).replace('_label.nrrd', '')] = np.sum(nda)/1000
```

```python
for key in lymph_node_sizes.keys():
    print(key, ': {0:.2f} cc'.format(lymph_node_sizes[key]))
```
